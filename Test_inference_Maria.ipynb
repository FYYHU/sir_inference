The EpidemicModel class implements the base model. Its method `time_evolution` runs the simulation
for given 
- recover_probas = $\mu_i(t)$ 
- transmissions = $\{i , j, \lambda_{ij}(t)\}$ (for i and j in contact at $t$)

The ProximityModel class corresponds to a simple model where
- recover_probas = $\mu$ is constant
- transmission rate = $\lambda$ is constant
- the contact network changes at each time step
- proba of contact = $\exp(-d(i,j)/s)$

# Sanity check starting with equal partition of the population

```python
#starts simulations with approximatelly the same number of inidividuals in each state
def generate_ground_truth(N):
    states=np.zeros(N)
    nb=np.zeros(3)
    frac=float(N)/3.0
    frac=int(frac)
    states[N-frac:]=2*np.ones(frac)
    states[N-2*frac-1:N-frac-1]=np.ones(frac)
    np.random.shuffle(states)
    return states
```

The probas of M observed individuals are set to be $P_S^i(0)=\delta(S-s_m[i])$ where $s_m[i]$ is the state of individual $i$ in the real world simulations

```python
def generate_initial_probas(M, states):
    freqs = [np.mean(states==s) for s in [0,1,2]]
    print("freqs = ", freqs)
    N = len(states)
    initial_probas = np.broadcast_to(freqs, (N, 3)).copy()
    observations = np.random.choice(N, M, replace=False)
    for i in observations:
        s = int(states[i])
        initial_probas[i] = np.zeros(3)
        initial_probas[i, s] = 1.
    return initial_probas
```

In order to make some sanity check, here we run a simulation where all the population is observed (the state is deterministic) and check that the predicted probas stay close to the reality for a considerable length of time

```python
# we observe the whole population persons at time t_obs=0
t_obs = 0
initial_probas = generate_initial_probas(M=N, states=model.states[t_obs])
infer = InferenceModel(initial_probas, model.x_pos, model.y_pos)
```

```python
infer.time_evolution(model.recover_probas, model.transmissions[t_obs:])
```

```python
interact(infer.plot_probas, t=IntSlider(min=0, max=30));
```

This is a routine that creates a 2D plot in which each individual occupies an (x,y) position at random. The colors of the dots indicate the probability of the individual to be in state S (respectively I and R). The red crosses correspond to the individuals that are in the same state given by model. 

```python
t_obs =0
t_fin=500
def compare_individuals(t):
    #marke in the scatter plot, the individuals that are really in that state by a red cross
    fig,ax = plt.subplots(1,3,figsize=(15,5),sharey=True)
    for s, state in enumerate("SIR"):
        ind = np.where(model.states[t+t_obs] == s)[0]
        #im=ax.scatter(infer.x_pos, infer.y_pos, c=infer.probas[t, :, s])
        im=ax[s].scatter(infer.x_pos, infer.y_pos, c=infer.probas[t, :, s],
                       cmap="Blues", vmin=0, vmax=1)
        ax[s].scatter(np.take(infer.x_pos,ind), np.take(infer.y_pos,ind),marker='+', vmin=0, vmax=1,color="Red")
        ax[s].set(title=state)
    fig.subplots_adjust(top=0.9)
    fig.suptitle("t="+str(t),fontsize=16)
    fig.colorbar(im)
    fig.tight_layout()
interact(compare_individuals, t=IntSlider(min=t_obs, max=50))
```

```python
compare_individuals(0)
compare_individuals(30)
compare_individuals(50)
compare_individuals(100)
```

# Many Simulations to do a scatter plot

Performs nsimu simulations starting with the same initial conditions (same positions and same initial states).
The scatter plot is then obtained as:
    x-axis: the average, taken over the nsimu simulations, of the probability of an individual to be in state s
    y-axis: the average number of time the individual was really in that state as predicted by the propagation               model: # times in state s/nsimu
Each dot represents an individual.
The inference is started by imposing that all individuals have probability one to be in the correct state.
The generative model starts with initial condition of having the same number of individuals in each state.

```python
from sir_model import ProximityModel
from inference_model import InferenceModel
#
#   N=population size
#   tVals=times at which the scatter plot should be plotted
#   nsimu=number of simulations over which the averages are taken
#   t_fin=final time of the simulations
#   lamb=given there was a contact, the transmission probability
#   mu=probability of recovery
#   scale=distace scale for the exponential random variable defining the probability of contact
#   t_obs=the time at which the system is seen
#
#
def get_scatter(N,tVals=[0,1,5,10,50,100],nsimu=100,t_fin=100,lamb=0.01,mu=0.01,scale=0.5):
    #initial conditions for the inference model to be exactly 1 at the correct state of each individual
    def generate_initial_probas(M, states):
        freqs = [np.mean(states==s) for s in [0,1,2]]
        print("freqs = ", freqs)
        N = len(states)
        initial_probas = np.broadcast_to(freqs, (N, 3)).copy()
        observations = np.random.choice(N, M, replace=False)
        for i in observations:
            s = int(states[i])
            initial_probas[i] = np.zeros(3)
            initial_probas[i, s] = 1.
        return initial_probas
    
    #starts simulations with approximatelly the same number of inidividuals in each state
    def generate_ground_truth(N):
        states=np.zeros(N)
        nb=np.zeros(3)
        frac=float(N)/3.0
        frac=int(frac)
        states[N-frac:]=2*np.ones(frac)
        states[N-2*frac-1:N-frac-1]=np.ones(frac)
        np.random.shuffle(states)
        return states
    
    x_pos = np.sqrt(N)*np.random.rand(N)
    y_pos = np.sqrt(N)*np.random.rand(N)
    ground_truth=generate_ground_truth(N)
    states_M=[]
    proba_infered=[]
    for n in range(nsimu):
        print(n)
        model = ProximityModel(N=N, scale=scale, mu=mu, lamb=lamb ,initial_states=ground_truth,x_pos=x_pos,y_pos=y_pos)
        model.run(tfin,print_every=500)
        states_M.append(model.states)
        print("expected number of contacts %.1f" % model.n_contacts)
        # we observe the whole population persons at time t_obs=0
        t_obs=0
        initial_probas = generate_initial_probas(M=N, states=model.states[t_obs])
        infer = InferenceModel(initial_probas, model.x_pos, model.y_pos)
        infer.time_evolution(model.recover_probas, model.transmissions[t_obs:],print_every=500)
        proba_infered.append(infer.probas)
    
    def get_averages(t,states_M,proba_infered,state=1):
        #computes the averages over the nsimu simulations of the probabilities of the inference model and of the number of times each individual was in state s
        av_M=np.zeros(N)
        av_I=np.zeros(N)
        for i in range(N-1):
            tot_M=0
            tot_I=0
            for n in range(nsimu):
                #state_M[n][t][i] should be the state of the reality at the n'th simulation at time t for inidividual i
                if states_M[n][t][i]==state:
                    tot_M+=1.0
                tot_I+=proba_infered[n][t, i , state]
            av_M[i]=tot_M/float(nsimu)
            av_I[i]=tot_I/float(nsimu)
        return av_M, av_I
    
    #plots a scatter plot as defined above 
    def get_scatter(tVals):
        fig,ax = plt.subplots(2,3,figsize=(15,2*5),sharey=True)
        i=0
        for row in range(2):
            for col in range(3):
                for s, state in enumerate("S"):
                    av_M, av_I=get_averages(tVals[i],states_M,proba_infered,state=s)
                    ax[row][col].scatter(av_I,av_M,color="red")
                    ax[row][col].set(title=state+" , t="+str(tVals[i]))
                    ax[row][col].set_xlabel("$P_"+state+"^i(t)$")
                    ax[row][col].set_ylabel("# occurance in s/nsimul")
                    ax[row][col].set_xlim([0, 1])
                    ax[row][col].set_ylim([0, 1])
                    ax[row][col].plot([0,1],[0,1],'--','red')
                    i+=1
            fig.tight_layout()
    get_scatter(tVals)
```

```python
get_scatter(N=500,tVals=[0,1,5,10,50,100],nsimu=100,t_fin=100,lamb=0.01,mu=0.01,scale=0.5)
```
